```python
X = df2.drop(["affairs"],axis=1).values
y = df2["affairs"].values
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
X_train
```

```python
X_test = sc.transform(X_test)
X_test
```

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(X_train,y_train)
```

```python
y_pred = log.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
grid_param = {
    "C": [0.1,0.25,0.5,1,1.5],
    "penalty" : ["l2"],
    "solver" : ["newton-cg","saga","lbfgs","sag"]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_log = GridSearchCV(estimator=log, param_grid=grid_param, n_jobs=-1, cv=10)
grid_log.fit(X_train,y_train)
```

```python
best_grid = grid_log.best_estimator_
```

```python
y_pred = best_grid.predict(X_test)
```

```python
y_pred
```

```python
accuracy_score(y_test,y_pred)
```
